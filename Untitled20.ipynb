```python
import numpy as np
import pandas as pd
import warnings



from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = pd.read_csv("/content/student-mat.csv")



X = data.drop(columns=["G3"])
y = data["G3"]



label_encoders = {}
for column in X.select_dtypes(include=["object"]).columns:
    label_encoders[column] = LabelEncoder()
    X[column] = label_encoders[column].fit_transform(X[column])

giris_boyutu =32  # Giriş katmanı boyutu
gizliKatman1_boyutu = 99  # İlk gizli katman boyutu
gizliKatman2_boyutu = 3  # İkinci gizli katman boyutu
cikis_boyutu = 1  # Çıktı katmanı boyutu
learning_rate = 0.1

W1=np.random.randn(gizliKatman1_boyutu, giris_boyutu)
W2=np.random.randn(gizliKatman2_boyutu, gizliKatman1_boyutu)
W3=np.random.randn(cikis_boyutu, gizliKatman2_boyutu)



def jacobian_sigmoid(x):
    x= np.float128(x)
    sigmoid= 1 / (1 + np.exp(-x))
    diag_values = sigmoid * (1 - sigmoid)
    return np.diag(diag_values)

warnings.filterwarnings('ignore')
for epoch in range(1000):
  for index,row in X.iterrows():
        # İleri besleme (Forward propagation)
        giris=list(row)
        hedef=y.get(index)
        z1 = np.dot(W1,giris)
        z1 = np.float128(z1)
        a1 = 1 / (1 + np.exp(-z1))  # Sigmoid aktivasyon fonksiyonu
        z2 = np.dot(W2,a1)
        a2 = 1 / (1 + np.exp(-z2))  # Sigmoid aktivasyon fonksiyonu
        z3 = np.dot(W3,a2)
        a3 = z3  # doğrusal
        # Hata hesaplama (Loss calculation)
        e = hedef-a3
        s3=-2*1*e

        s2=np.dot(jacobian_sigmoid(a2),np.transpose(W3))*s3
        #print(jacobian_sigmoid(a1))
        s1 = np.dot(W2.T, s2) * jacobian_sigmoid(z1).T
        W3=W3-learning_rate*s3*a2
        W2=W2-learning_rate*s2*a1
        W1 = W1 - learning_rate * np.outer(s1, giris).sum(axis=0)


        # Geriye yayılma (Backpropagation)

hata = []
for index, row in X.iterrows():
    giris=list(row)
    hedef=y.get(index)
    z1 = np.dot(W1,giris)
    z1 = np.float128(z1)
    a1 = 1 / (1 + np.exp(-z1))  # Sigmoid aktivasyon fonksiyonu
    z2 = np.dot(W2,a1)
    a2 = 1 / (1 + np.exp(-z2))  # Sigmoid aktivasyon fonksiyonu
    z3 = np.dot(W3,a2)
    a3 = z3  # doğrusal
    # Hata hesaplama (Loss calculation)
    e = hedef-a3[0]
    hata.append(e)

print(hata)
```

```python
min_val = np.min(hata)
max_val = np.max(hata)

num_bins = int((max_val - min_val) / 2) + 1
bins = [min_val + 2 * i for i in range(num_bins + 1)]

frequency = np.histogram(hata, bins=bins)[0]

for i in range(num_bins):
    print(f"{bins[i]:.2f} - {bins[i+1]:.2f}: {frequency[i]}")
```
