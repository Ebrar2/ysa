```python
import numpy as np
import pandas as pd
import warnings
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt



from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest

data = pd.read_csv("student-mat.csv")



X = data.drop(columns=["G3"])
y = data["G3"]

tumVeriSeti=data

label_encoders = {}
for column in X.select_dtypes(include=["object"]).columns:
    label_encoders[column] = LabelEncoder()
    X[column] = label_encoders[column].fit_transform(X[column])+1

def jacobian_sigmoid(x):
    x= np.float128(x)
    diag_values = x * (1 - x)
    return np.diag(diag_values)

class agirliklar:
  def __init__(self, W1, W2,W3):
    self.W1 = W1
    self.W2 = W2
    self.W3=W3
warnings.filterwarnings('ignore')
def egitim(epoch_size,veri_seti,W1,W2,W3):
  for epoch in range(epoch_size):
    for index,row in veri_seti.iterrows():
          giris=list(row)
          hedef=y.get(index)
          z1 = np.dot(W1,giris)
          z1 = np.float128(z1)
          a1 = 1 / (1 + np.exp(-z1))
          z2 = np.dot(W2,a1)
          z2 = np.float128(z2)
          a2=1 / (1 + np.exp(-z2))
          z3=np.dot(W3,a2)
          a3=z3
          e = hedef-a3
          s3=-2*1*e
          s2=np.dot(jacobian_sigmoid(a2),np.transpose(W3))*s3
          s1=np.dot(np.dot(jacobian_sigmoid(a1),np.transpose(W2)),s2)
          W3=W3-learning_rate*s3*np.transpose(a2)
          W2=W2-learning_rate*s2*np.transpose(a1)
          W1=W1-learning_rate*s1*np.transpose(giris)
  return agirliklar(W1,W2,W3)

def hata_hesabi(veri_seti,W1,W2,W3):
  hata=[]
  karesel_hata=0
  for index,row in veri_seti.iterrows():
          giris=list(row)
          hedef=y.get(index)
          z1 = np.dot(W1,giris)
          z1 = np.float128(z1)
          a1 = 1 / (1 + np.exp(-z1))
          z2 = np.dot(W2,a1)
          z2 = np.float128(z2)
          a2=1 / (1 + np.exp(-z2))
          z3=np.dot(W3,a2)
          a3=z3
          e = hedef-a3
          karesel_hata+=pow(e[0],2)
          hata.append(e[0])
  print('Karesel Hata Toplami:'+str(karesel_hata))
  return hata



def histogram_olustur(hata,title):
  plt.hist(hata, bins=10, color='skyblue', edgecolor='black')
  plt.title(title)
  plt.xlabel('Hata')
  plt.ylabel('Frekans')
  plt.show()


giris_boyutu =len(X.columns)
gizliKatman_boyutu1 = 99
gizliKatman_boyutu2 = 3
cikis_boyutu = 1
learning_rate = 0.1

genel_W1=np.random.randn(gizliKatman_boyutu1,giris_boyutu)
genel_W2=np.random.randn(gizliKatman_boyutu2,gizliKatman_boyutu1)
genel_W3=np.random.randn(cikis_boyutu,gizliKatman_boyutu2)

#Eğitim 10 epoch
sonuc=egitim(10,X,genel_W1,genel_W2,genel_W3)
hata1=hata_hesabi(X,sonuc.W1,sonuc.W2,sonuc.W3)
histogram_olustur(hata1,'32-99-3-1:epoch:10:a=0.1')

#Eğitim 20 epoch
sonuc=egitim(20,X,genel_W1,genel_W2,genel_W3)
hata2=hata_hesabi(X,sonuc.W1,sonuc.W2,sonuc.W3)
histogram_olustur(hata2,'32-99-3-1:epoch:20:a=0.1')

#Eğitim 30 epoch
sonuc=egitim(30,X,genel_W1,genel_W2,genel_W3)
hata3=hata_hesabi(X,sonuc.W1,sonuc.W2,sonuc.W3)
histogram_olustur(hata3,'32-99-3-1:epoch:30:a=0.1')
```
