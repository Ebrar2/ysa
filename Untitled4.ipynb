```python
import numpy as np
import pandas as pd
import warnings



from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = pd.read_csv("student-mat.csv")



X = data.drop(columns=["G3"])
y = data["G3"]



label_encoders = {}
for column in X.select_dtypes(include=["object"]).columns:
    label_encoders[column] = LabelEncoder()
    X[column] = label_encoders[column].fit_transform(X[column])


giris_boyutu =32   # Giriş katmanı boyutu
gizliKatman_boyutu = 3    # Gizli katman boyutu
cikis_boyutu = 1    # Çıktı katmanı boyutu
learning_rate = 0.1

W1=np.random.randn(gizliKatman_boyutu,giris_boyutu)
W2=np.random.randn(cikis_boyutu,gizliKatman_boyutu)



def jacobian_sigmoid(x):
    x= np.float128(x)
    sigmoid= 1 / (1 + np.exp(-x))
    diag_values = sigmoid * (1 - sigmoid)
    return np.diag(diag_values)

warnings.filterwarnings('ignore')
for epoch in range(10):
  for index,row in X.iterrows():
        # İleri besleme (Forward propagation)
        giris=list(row)
        hedef=y.get(index)
        z1 = np.dot(W1,giris)
        z1 = np.float128(z1)
        a1 = 1 / (1 + np.exp(-z1))  # Sigmoid aktivasyon fonksiyonu
        z2 = np.dot(W2,a1)
        a2=z2    #doğrusal
        # Hata hesaplama (Loss calculation)
        e = hedef-a2
        s2=-2*1*e
        s1= np.dot(jacobian_sigmoid(a1),np.transpose(W2))*s2
        W2=W2-learning_rate*s2*a1
        W1=W1-learning_rate*s1*giris
        # Geriye yayılma (Backpropagation)


hata=[]



for index,row in X.iterrows():
   giris=list(row)
   hedef=y.get(index)
   z1 = np.dot(W1,giris)
   z1 = np.float128(z1)
   a1 = 1 / (1 + np.exp(-z1))  # Sigmoid aktivasyon fonksiyonu
   z2 = np.dot(W2,a1)
   #a2 = 1 / (1 + np.exp(-z2))
   a2=z2    #doğrusal
   # Hata hesaplama (Loss calculation)
   e = hedef-a2[0]
   hata.append(e)

print(hata)
```
